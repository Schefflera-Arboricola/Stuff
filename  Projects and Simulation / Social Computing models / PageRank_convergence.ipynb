```python
import numpy as np
import random

a,b,c,d=random.randint(-100,100),random.randint(-100,100),random.randint(-100,100),random.randint(-100,100)
s1=str(a)+' '+str(b)+';'+str(c)+' '+str(d)
A=np.mat(s1)
x,y=random.randint(-100,100),random.randint(-100,100)
s2=str(x)+';'+str(y)
v=np.mat(s2)
for i in range(2000):
    z=A*v
    z=z/np.linalg.norm(z)
    v=z
    print(v)
```

**Conclusion** : "applying the matrix, normalize and repeat" makes any random vector converge to normalized eigenvector of the matrix.
